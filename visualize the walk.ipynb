# Visualize the walk

```python
# Let's visualize this random walk! Remember how you could use matplotlib to build a line plot?

import matplotlib.pyplot as plt
x = [1,2,3,4,5] 
y = [4,6,4,2,6]
plt.plot(x, y)
plt.show()

# The first list you pass is mapped onto the x axis and the second list is mapped onto the y axis.
```

```python
# Initialization
import numpy as np
np.random.seed(123)
random_walk = [0]

for x in range(100):
    step = random_walk[-1]
    dict = np.random.randint(1,7)
    
    if dict <=2:
        step = max(0, step-1)
    elif dict <=5:
        step=step+1
    else:
        step = step+np.random.randint(1,7)
    
    random_walk.append(step)    
```

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```
